Droping Unwanted Columns

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

Droping Duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

Cleaning Rate Column

```python
df['rate'].unique()
```

Removing 'NEW','nan','-' and '/' values from rate column

```python
def new_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate']= df['rate'].apply(new_rate)
df['rate'].head()
```

Fill null values

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
df['rate'].isnull().sum()
df['rate'].head()
```

Droping null values

```python
df = df.dropna()
df.head()
```

checking "approx_cost(for two people)" column for clining

```python
df['approx_cost(for two people)'].unique()
```

```python
def remove_comma(value):
    value=str(value)
    if ',' in value:
        value= value.replace(',','')
        return float(value)
    else:
        return float(value)
df['approx_cost(for two people)'].apply(remove_comma).unique()
```

Top 10 Highest Rating Restaurants

```python
highest_rating = df.sort_values(by=['rate'],ascending=False)
highest_rating.head(10)
```

Top 10 Lowest Rating Restaurants

```python
lowest_rating = df.sort_values(by=['rate'],ascending=True)
lowest_rating.head(10)
```

Restaurants delivering Online or not


```python
plt.figure(figsize=(6,4)) 
online_counts = df['online_order'].value_counts().plot(kind='bar',rot=0,color = sns.color_palette())
plt.title("Restaurants Delivering Online or Not")
plt.ylabel('Restaurants')
plt.xlabel('Online Delivery')
```

Restaurants allowing table booking or not

```python
plt.figure(figsize=(6,4)) 
online_counts = df['book_table'].value_counts().plot(kind='bar',rot=0,color = sns.color_palette())
plt.title("Restaurants Table Booking or Not")
plt.ylabel('Restaurants')
plt.xlabel('Table Booking')
```

Find Best Locations

```python
locations = df.groupby('location')[['rate','votes','listed_in(city)']].apply(lambda x: x.nlargest(10,columns = 'rate')).droplevel(1).reset_index()

locations.set_index('location')
```

Top Ten cities by Ratings

```python
# Grouping location name column 
Top_cities=pd.DataFrame(df.groupby('location').sum()['rate'])
# Sorting location   column
Top_cities=Top_cities.sort_values('rate',ascending=False)
Top_cities.head(10)
```

Relation between Location and Rating

```python
plt.figure(figsize=(6,4)) 
sns.relplot(data=df.head(1000), x="rate", y="location",hue ='rate',kind ="scatter")
```

Types of Restaurants

```python
Types= df['rest_type'].value_counts().head(10).reset_index()
Types.index +=1
Types
```

```python
plt.figure(figsize=(6,4)) 
sns.barplot(data= Types,y='index',x = 'rest_type')
plt.title('Types of Restaurants ', fontsize = 15)
plt.xlabel(' Type counts', fontsize = 10)
plt.ylabel('Categories', fontsize = 10)
```

```python
types_of_service = df['listed_in(type)'].value_counts().reset_index()
types_of_service.index +=1
types_of_service
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x = df['listed_in(type)'])
plt.title('Types of Services', fontsize = 15)
plt.xlabel('Services', fontsize = 10)
plt.ylabel('Counts', fontsize = 10)
plt.xticks(rotation=45)
plt.show()
```

Top 10 Restaurant by Ratings and votes

```python
top_res=pd.DataFrame(df.groupby(['name','rate']).sum()['votes'])
top_res=top_res.sort_values('votes',ascending=False)
top_res.head(10)
```

Top 10 Restaurant by Outlets

```python
plt.figure(figsize=(6,4)) 
df['name'].value_counts()[:10].plot(kind = 'bar',rot=30, color = sns.color_palette("Paired"))
plt.title('Top 10 Restaurant ', fontsize = 15)
plt.xlabel('Restaurant Name ', fontsize = 10)
plt.ylabel('No. of Outlets', fontsize = 10)
```
